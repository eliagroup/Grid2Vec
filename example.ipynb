```python
from grid2vec.grid2op_compat import load_grid_grid2op, all_actions_action_dump
from grid2vec.env import make_env, vector_reset, vector_step, get_reward, get_done
from grid2vec.solver_interface import compute_obs
import grid2op
import numba  # TODO find out why pandapower can't use numba
import jax
import jax.numpy as jnp
```

```python
grid = load_grid_grid2op("l2rpn_case14_sandbox", include_chronic_indices=[0, 1, 2, 3])
```

```python
grid.net.load
```

```python
action_dump = all_actions_action_dump(grid2op.make("l2rpn_case14_sandbox"))
```

```python
random_key = jax.random.PRNGKey(0)
batch_size = 4
```

```python
env = make_env(grid, batch_size)
```

```python
env.topo_vect
```

```python
env = make_env(grid, batch_size)
env = vector_reset(env)
obs = compute_obs(env)

for _ in range(10):
    action = jax.random.choice(random_key, len(action_dump), shape=(batch_size,))
    env = vector_step(env, **action_dump.actions[action].asdict())
    obs = compute_obs(env)
    print(get_reward(env.grid, obs))
    vector_reset(env, which=get_done(obs))
```
